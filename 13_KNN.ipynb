# KNN k - nearest negihbore 

```python
import pandas as pd
```

```python
df = pd.read_csv("D:/ETL HIVE/PYTHON/ADV PY/Cars93.csv")
```

```python
df
```

```python
for i in df.columns:
    if df[i].dtypes==object:
        df[i]=df[i].fillna(df[i].mode()[0])
    else:
        df[i]=df[i].fillna(df[i].mean())
```

```python
df.isna().sum()
```

```python
X = df.drop('Origin',axis=1)
```

```python
Y  = df[['Origin']] 
```

```python
cat = []
con = []

for i in X.columns:
    if X[i].dtypes==object:
        cat.append(i)
    else:
        con.append(i)
```

```python
Xcat = X [cat]
Xcon = X [con]
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
ss = StandardScaler()
```

```python
le = LabelEncoder()
```

```python
for i in Xcat.columns:
    Xcat[i]=le.fit_transform(Xcat[i])
```

```python
Xcon = pd.DataFrame(ss.fit_transform(Xcon),columns=con)
```

```python
X = Xcon.join(Xcat)
```

```python
out = []

for i in Xcon.columns:
    o = Xcon[(Xcon[i] < -3 ) | (Xcon[i] > 3)].index
    out.extend(o)
```

```python
a = list(set(out))
a
```

```python
X = X.drop(index=a,axis=0)
Y = Y.drop(index=a,axis=0)
```

```python
X.index=range(X.shape[0])
Y.index=range(Y.shape[0])
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knr = KNeighborsClassifier(n_neighbors=8)
```

```python
model = knr.fit(xtrain,ytrain)
```

```python
tr_pred = model.predict(xtrain)
ts_pred = model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
tr_score = accuracy_score(ytrain,tr_pred)
tr_score
```

```python
ts_score = accuracy_score(ytest,ts_pred)
ts_score
```
