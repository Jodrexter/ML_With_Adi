```python
import pandas as pd
```

```python
df = pd.read_csv("D:/ETL HIVE/PYTHON/ADV PY/Cars93.csv")
```

```python
df
```

```python
df.isna().sum().plot(kind='bar')
```

```python
for i in df.columns:
    if df[i].dtypes==object:
        df[i]=df[i].fillna(df[i].mode()[0])
    else:
        df[i]=df[i].fillna(df[i].mean())
```

```python
df.isna().sum().plot(kind='bar')
```

```python
X = df.drop('Horsepower',axis=1)
```

```python
X
```

```python
Y = df[['Horsepower']]
Y
```

```python
cat = []
con = []

for i in X.columns:
    if X[i].dtypes==object:
        cat.append(i)
    else:
        con.append(i)
```

```python
Xcat = X [cat]
Xcon = X [con]
```

```python
Xcat
```

```python
Xcon
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
le = LabelEncoder()
ss = StandardScaler()
```

```python
for i in Xcat.columns:
    Xcat[i]=le.fit_transform(Xcat[i])
```

```python




























































Xcat
```

```python
Xcon = pd.DataFrame(ss.fit_transform(Xcon),columns=con)
```

```python
Xcon
```

```python
X = Xcon.join(Xcat)
X
```

```python
out = []
for i in Xcon.columns:
    o = Xcon[(Xcon[i]<-3)|(Xcon[i]>3)].index
    out.extend(o)
```

```python
out
```

```python
a = list(set(out))
```

```python
a
```

```python
X = X.drop(index=a,axis=0)
Y = Y.drop(index=a,axis=0)
```
